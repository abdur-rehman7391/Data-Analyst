```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
```

```python
data = '/content/jena_climate_2009_2016.csv'
```

```python
df = pd.read_csv(data)
df.head(5)
```

```python
df.shape
```

```python
df = df[5::6]
df.head()
```

```python
df.index = pd.to_datetime(df['Date Time'], format = '%d.%m.%Y %H:%M:%S')
df[:26]
```

```python
df.shape
```

```python
temp = df['T (degC)']
temp.plot()
```

```python
def df_to_X_y(df, window_size = 5):
    df_as_np = df.to_numpy()
    X = []
    y = []
    for i in range(len(df_as_np)-window_size):
        row = [[a] for a in df_as_np[i: i+5]]
        X.append(row)
        label = df_as_np[i+5]
        y.append(label)
    return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X, y = df_to_X_y(temp, WINDOW_SIZE)
X.shape, y.shape
```

```python
X_train, y_train = X[:60000], y[:60000]
X_val, y_val = X[60000:65000], y[60000:65000]
X_test, y_test = X[65000:], y[65000:]
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
model1 = Sequential()
model1.add(InputLayer((5,1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
```

```python
from tensorflow.keras.metrics import RootMeanSquaredError
cp = ModelCheckpoint('model.keras', save_best_only = True)
#model1.compile(loss = MeanSquareError(), optimizer = Adam(learning_rate = 0.0001), metrics= [RootMeanSqaureError()])

model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
#model1.fit(X_train, y_test, validation_data=(X_val, y_val), epochs= 10, callbacks=[cp])
model1.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('model.keras')
```

```python
train_pred = model1.predict(X_train).flatten()
train_result = pd.DataFrame(data={'Train Predict: ': train_pred, 'Actual': y_train})
train_result
```

```python
import matplotlib.pyplot as plt
plt.plot(train_result['Train Predict: '][:10])
plt.plot(train_result['Actual'][:10])
```

```python
val_pred = model1.predict(X_val).flatten()
val_result = pd.DataFrame(data={'Val Predict: ': val_pred, 'Actual': y_val})
val_result
```

```python

plt.plot(val_result['Val Predict: '][:100])
plt.plot(val_result['Actual'][:100])
```

```python
test_pred = model1.predict(X_test).flatten()
test_result = pd.DataFrame(data={'Test Predict: ': test_pred, 'Actual': y_test})
test_result
```

```python
plt.plot(test_result['Test Predict: '][:100])
plt.plot(test_result['Actual'][:100])
```
