```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score,  classification_report
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv(r"E:\Study\Data Analyst\Practice data\CyberBullying Detection Dataset\final_hateXplain.csv")
df.head(9)
```

```python
categorical_cols = ['Race', 'Gender', 'label','Religion','Sexual_Orientation','Miscellaneous']
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
scaler = StandardScaler()

# Fit and transform the numerical data
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df
```

```python
vectorizer = TfidfVectorizer()
X_text = vectorizer.fit_transform(df['comment'])
```

```python
X = df.drop(columns=['comment', 'label'])  # Potentially exclude unnecessary columns
y = df['label']
```

```python
X_numerical = scaler.fit_transform(X)
```

```python
X_numerical
```

```python
print(X_numerical.shape)
print(X_text.shape)
```

```python
X_numerical.reshape(-1, 1)
X_text = X_text.toarray()
```

```python
X = np.column_stack((X_numerical, X_text))
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
 random_state=42)
```

```python
svm_model = SVC()  # Create an SVC object
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```
