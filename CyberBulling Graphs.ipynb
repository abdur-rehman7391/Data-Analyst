```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df0 = pd.read_csv('/content/final_hateXplain.csv')
```

```python
df0.head(2)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df0, x='label')
plt.title('Distribution of Labels')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df0, x='Race')
plt.title('Distribution of Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()
```
